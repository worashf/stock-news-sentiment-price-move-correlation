```python
import sys
from src import DataLoader
from src.analysis_pipeline import TechnicalAnalysisPipeline

# Add src directory to path
sys.path.append('../../src')

try:
    # Initialize data loader
    data_loader = DataLoader()

    # Load stock data
    tickers = ['AAPL', 'MSFT', 'GOOG','AMZN','NVDA','TSLA','META']
    stock_data = data_loader.load_multiple_stocks(tickers)

    # Initialize analysis pipeline
    analysis_pipeline = TechnicalAnalysisPipeline()

    # Option 1: Full analysis with all indicators
    full_results = analysis_pipeline.analyze_multiple_stocks(stock_data)
    print(f"full_results,{full_results}")

    # Display AAPL results
    aapl_result = full_results['AAPL']
    apple_result = full_results['MSFT']
    print('AAPL:', aapl_result)
    aapl_result['figure'].show()
    print("AAPL Financial Metrics:")
    print(aapl_result['metrics'])

    # Display correlation matrix if available
    if 'correlation' in full_results:
        full_results['correlation'].show()

    # Option 2: Custom analysis with specific indicators
    custom_indicators = ['RSI', 'MACD', 'OBV', 'ATR']
    custom_result = analysis_pipeline.analyze_custom_indicators(
        stock_data['AAPL'], 'AAPL', custom_indicators
    )
    custom_result['figure'].show()

except Exception as e:
    print(f"Analysis failed: {str(e)}")
```
